## Preprocessing

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk_stopwords = set(stopwords.words("english"))
my_stopwords = ['beer', 'look', 'smell', 'taste', 'feel', 'mouthfeel', 'overall', 'good', 'terrible', 'well', 'korean']
new_stopwords = nltk_stopwords.union(my_stopwords)
```

```python
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
from nltk import pos_tag
from nltk.corpus import wordnet
```

```python
def get_pos(w):
    tag = pos_tag([w])[0][1][0].upper()
    if tag == "V":
        return wordnet.VERB
    elif tag == "N":
        return wordnet.NOUN
    elif tag == "J":
        return wordnet.ADJ
    elif tag == "R":
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def my_tokenizer(r):
    r1 = word_tokenize(r)
    r2 = [w.lower() for w in r1 if w.isalpha()]
    r3 = [w for w in r2 if not w in new_stopwords]
    r4 = [lem.lemmatize(w, get_pos(w)) for w in r3]
    return r4
```

# Clustering

### top15

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(tokenizer = my_tokenizer)
tfidf_top = tfidf.fit_transform(top2)
tfidf_terms = tfidf.get_feature_names()

tfidf_top_doc = pd.DataFrame(tfidf_top.toarray(), columns=tfidf_terms)
tfidf_top_doc.index = top2
tfidf_top_doc
```

```python
from sklearn.cluster import KMeans
```

```python
distance = []

K = range(1, 10)
for k in K:
    km1 = KMeans(n_clusters=k, random_state=0)
    km2 = km1.fit(tfidf_top)
    d = km2.inertia_
    distance.append(d)
    
distance
```

```python
import matplotlib.pyplot as plt
plt.plot(K, distance)
plt.show()
```

```python
km = KMeans(n_clusters=3, random_state=0)
km.fit(tfidf_top)
```

```python
oc = km.cluster_centers_.argsort()[:, ::-1]

for i in range(0, 3):
    top_words = [tfidf_terms[x] for x in oc[i, :5]]
    print(top_words)
```
